```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
# Gr√°ficos
```

```python
tips.head()
```

```python
len(tips) #244 filas 
```

```python
sns.set_style('darkgrid') # darkgrid, whitegrid, dark, white, ticks
```

```python
sns.distplot(tips['total_bill'], # variable de an√°lisis
             kde = False, # kernel distribution
            bins=20 # particiones
            )
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, 
              kind = 'kde') #reg
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, 
              kind = 'reg') 
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, 
              kind = 'hex') 
```

```python
# Grafico para combiar todas las variables num√©ricas
sns.pairplot(tips)
```

```python
# Gr√°ficos para las variables categoricas
sns.pairplot(tips, 
             hue='sex', # diferenciaci√≥n por una variable
             palette = 'coolwarm' # temas de los colores
            )
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde = False)
```

```python
# kde Estimaci√≥n de la densidad de kernel
sns.kdeplot(tips['total_bill'])
```
