```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
# Boxplots nos da información acerca de la distribución de los datos
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# Nos da la posibilidad de ver todos los puntos que hacen referencia a cada fila del dataset
# Quizás con datasets de muchos datos no puede ser totalmente útil
# Su forma nos recuerda al gráfico de violín
sns.swarmplot(x = 'day', y = 'total_bill', data=tips)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data=tips, color = 'black')
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.factorplot(x = 'day', y='total_bill', data=tips, kind='bar')
```
