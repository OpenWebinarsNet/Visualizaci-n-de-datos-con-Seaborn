**Escribe el código necesario para crear los gráficos que se muestran a continuación. La mayoría de gráficos se pueden hacer con usolo una o dos líneas de código utilizando la librería de seaborn.**

```python
# Pon tu código aquí
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
# Pon tu código aquí
sns.distplot(titanic['fare'], kde=False, color = 'red', bins=25)
```

```python
# Pon tu código aquí
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
# Pon tu código aquí
sns.swarmplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
# Pon tu código aquí
sns.countplot(x='sex',data=titanic, palette='Set2')
```

```python
# Pon tu código aquí
sns.heatmap(titanic.corr(), cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
# Pon tu código aquí
g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist,'age')
```
