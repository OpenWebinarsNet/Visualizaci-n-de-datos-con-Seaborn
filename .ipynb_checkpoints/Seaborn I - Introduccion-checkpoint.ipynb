```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
# Gráficos
```

```python
tips.head()
```

```python
len(tips) #244 filas 
```

```python
sns.set_style('darkgrid') # darkgrid, whitegrid, dark, white, ticks
```

```python
sns.distplot(tips['total_bill'], # variable de análisis
             kde = False, # kernel distribution
            bins=20 # particiones
            )
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, 
              kind = 'kde') #reg
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, 
              kind = 'reg') 
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, 
              kind = 'hex') 
```

```python
# Grafico para combiar todas las variables numéricas
sns.pairplot(tips)
```

```python
# Gráficos para las variables categoricas
sns.pairplot(tips, 
             hue='sex', # diferenciación por una variable
             palette = 'coolwarm' # temas de los colores
            )
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde = False)
```

```python
# kde Estimación de la densidad de kernel
sns.kdeplot(tips['total_bill'])
```
