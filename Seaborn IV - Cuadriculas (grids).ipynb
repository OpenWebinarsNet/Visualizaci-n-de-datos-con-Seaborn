```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid') # darkgrid, whitegrid, dark, white, ticks
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot) # histogramas 
g.map_upper(plt.scatter) # gráficos de puntos (gráfico de dispersión)
g.map_lower(sns.kdeplot) # estimación de la distribución de Kernel
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Creará una columa por cada categoría de time 
# Creará una fila por cada categoría de smoker
sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
t = sns.FacetGrid(data=tips, col='time', row='smoker')
t.map(sns.distplot, 'total_bill')
```

```python
t = sns.FacetGrid(data=tips, col='time', row='smoker')
t.map(plt.scatter, 'total_bill', 'tip')
```
